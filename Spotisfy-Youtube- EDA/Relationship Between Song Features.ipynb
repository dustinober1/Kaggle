<b>GOAL #1:</b> Analyze the relationship between the features of the songs, such as danceability, energy, valence, and instrumentalness, and the number of streams and views on YouTube.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('Spotify_Youtube.csv')
```

```python
# Select the features and the target variables
features = ['Danceability', 'Energy', 'Valence', 'Instrumentalness']
target_streams = 'Stream'
target_views = 'Views'
```

```python
# Create a pairplot to visualize the relationship between the features and the targets
sns.pairplot(df, x_vars=features, y_vars=target_streams, kind='reg')
sns.pairplot(df, x_vars=features, y_vars=target_views, kind='reg')
```

```python
# Calculate the correlation matrix
corr_streams = df[features + [target_streams]].corr()
corr_views = df[features + [target_views]].corr()
```

```python
# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_streams, annot=True, cmap='coolwarm')
plt.title('Correlation matrix of features and streams')

plt.figure(figsize=(10,8))
sns.heatmap(corr_views, annot=True, cmap='coolwarm')
plt.title('Correlation matrix of features and views')

plt.show()
```
