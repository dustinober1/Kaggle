```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
```

```python
df.head(2)
```

```python
df.info()
```

```python
df = df.drop(['Title', 'Url_spotify', 'Uri', 'Description', 'official_video', 'Channel', 'Url_youtube'], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df['Views']
```

```python
df.groupby('Artist').sum()['Views'].sort_values(ascending = False).head(10)
```

```python
# group the DataFrame by Artist and sum the Views, then sort the results in descending order and select the top 10
artists = df.groupby('Artist')['Views'].sum().sort_values(ascending=False).head(10)

# sort the top 10 artists in ascending order
artists = artists.sort_values(ascending=True)

# convert the Series back into a DataFrame
artists = pd.DataFrame({'Artist': artists.index, 'Views': artists.values})

# create the line plot using seaborn
sns.set_style('darkgrid')
plt.figure(figsize=(24,12))
sns.lineplot(data=artists, x='Artist', y='Views', color='purple')
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.title('The Top Streamed Artists on YouTube Music', fontsize=20)
plt.xlabel('Artist', fontsize=16)
plt.ylabel('Total Views', fontsize=16)
plt.show()
```

```python
df.groupby('Artist').sum()['Stream'].sort_values(ascending = False).head(10)
```

```python
# group the DataFrame by Artist and sum the Stream column, then sort the results in descending order and select the top 10
artists2 = df.groupby('Artist')['Stream'].sum().sort_values(ascending=False).head(10)

# sort the top 10 artists in ascending order
artists2 = artists2.sort_values(ascending=True)

# convert the Series back into a DataFrame
artists2 = pd.DataFrame({'Artist': artists2.index, 'Stream': artists2.values})

# create the line plot using seaborn
sns.set_style('darkgrid')
plt.figure(figsize=(24,12))
sns.lineplot(data=artists2, x='Artist', y='Stream', color='green')
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.title('The Top Streamed Artists on Spotify', fontsize=20)
plt.xlabel('Artist', fontsize=16)
plt.ylabel('Total Streams', fontsize=16)
plt.show()
```

```python
df.sort_values(by = 'Views', ascending = False)[['Track', 'Artist', 'Views']].head(10)
```

```python
holdi = df.sort_values(by = 'Views', ascending = False)[['Track', 'Artist', 'Views']].head(7)
plt.figure(figsize=(19,6))
sns.barplot(data = holdi, y= 'Track', x='Views', palette = 'Reds_r')
plt.title("Top 5 Songs on Youtube")
plt.xlabel("Billions of Views", color = 'lightgrey')
plt.ylabel("Song", color = 'lightgrey')
plt.show()
```

```python
holdi = df.sort_values(by = 'Views', ascending = False)[['Track', 'Artist', 'Views']].head(7)
plt.figure(figsize=(19,6))
sns.scatterplot(data = holdi, x= 'Track', y='Views', hue = 'Artist', s = 150, palette = 'Reds_r')
plt.title("Top 5 Songs on Youtube")
plt.ylabel("Billions of Views", color = 'lightgrey')
plt.xlabel("Song", color = 'lightgrey')
plt.legend(loc='upper right')
plt.show()
```

```python
holdi = df.sort_values(by = 'Views', ascending = False)[['Track', 'Artist', 'Views']].head(7)
pivot_holdi = holdi.pivot_table(index='Track', columns='Artist', values='Views')
plt.figure(figsize=(19,6))
sns.heatmap(pivot_holdi, cmap = 'Reds', annot = True, fmt=".1f")
plt.title("Top 5 Songs on Youtube")
plt.ylabel("Song", color = 'lightgrey')
plt.xlabel("Artist", color = 'lightgrey')
plt.show()
```

```python
holdi = df.sort_values(by = 'Stream', ascending = False)[['Track', 'Artist', 'Stream']].head(6)
plt.figure(figsize=(19,6))
sns.barplot(data = holdi, x= 'Track', y='Stream', palette = 'Greens_r')
plt.title("Top 5 Songs on Spotify")
plt.ylabel("Billions of Streams", color = 'lightgrey')
plt.xlabel("Song", color = 'lightgrey')
plt.show()
```

```python
df['seconds'] = df['Duration_ms'].apply(lambda x: x/1000)
df3 = df['seconds'].sort_values(ascending = True).head(19000)
plt.figure(figsize=(19,6))
plt.title("Song Length Distribution (seconds)", loc = 'left', color = 'grey')
plt.ylabel([])
sns.histplot(df3, bins = 50, color = 'lightgrey')
```
